## Lade Libraries und Daten

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import warnings
from sklearn.model_selection import train_test_split
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)

pd.options.mode.chained_assignment = None

DF = pd.read_csv('data/titanic_enh.csv')
```

```python
DF
```

### Lade nötige Library/Libraries für KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

### Trainiere KNN Classifier mit Trainingsdaten

```python
df = DF.copy()
k = 7
# Split data into training and test set
train, test = train_test_split(df, test_size=0.2)

# Create Input- and Output-Structure and train Classifer
training_columns = ['Pclass', 'Sex', 'Age','Fare','Embarked', 'Has_Cabin', 'FamilySize', 'IsAlone', 'Title']

X = train[training_columns]
y = train[['Survived']]



# Instantiate knn model
model = KNeighborsClassifier(k)

# Train the model using the training sets
model.fit(X,y)
```

### Treffe Vorhersagen für Testdaten und berechne Performance-Metrik

```python
# Calculate predictions for test set

X = test[training_columns]
y = test[['Survived']]
y['Prediction'] = model.predict(X)

# Calculate Performance Metric
y['match'] = y['Survived'] == y['Prediction']
performance =  sum(y['match']) / len(y)
print('Performance: {:.1%}'.format(performance))
```

### Suche Hyperparameter

```python
k_value = []
value_performance = []
tries = 5
max_performance = 0
max_performance_value = 0

for i in range(1, 50):
    value = i
    overall_perf = 0
    for j in range(tries):
        
        model = KNeighborsClassifier(max_value = value, criterion ='entropy')
        #clf = DecisionTreeClassifier(max_depth = depth, criterion='entropy')
        
        df = DF.copy()
        train, test = train_test_split(df, test_size=0.2)

        
        training_columns = ['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'Has_Cabin', 'FamilySize', 'IsAlone', 'Title']

        X = train[training_columns]
        y = train[['Survived']]
        model.fit(X,y)

        
        X = test[training_columns]
        y = test[['Survived']]
        y['Prediction'] = model.predict(X)

        
        y['match'] = y['Survived'] == y['Prediction']
        performance =  sum(y['match']) / len(y)
        #rint('Performance: {:.1%}'.format(performance))
        overall_perf = overall_perf + performance
        
    k_value.append(value)
    value_performance.append(overall_perf/tries)
    if overall_perf/tries > max_performance:
        max_performance = overall_perf/tries
        max_performance_depth = value


print('Die höchste Performance von {:.1%} wurde mit einem k von {} erreicht.'.format(max_performance, max_performance_value))
g = sns.lineplot(x=k_value, y=value_performance)
g.axes.set_ylim(0.5,1)
g
```
