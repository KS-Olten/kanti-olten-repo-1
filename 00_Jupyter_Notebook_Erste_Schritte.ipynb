## 1. Was ist Jupyter Notebooks und Binder

**Jupyter Notebooks** ist eine weitere Ausführumgebung für Python. Jupyter Notebooks bestehen aus zwei Teilen:

> 1. Einem **Notebook**. Das Notebook ist eine Datei mit der Endung .ipynb. In das Notebook schreibst du deinen Code.
> 2. Einem **Jupyter Server**. Dieser Server kann mit einer Web-Adresse aufgerufen werden und führt dein Notebook im Browser aus. Wir verwenden dafür **Binder**.

**Binder** (betrieben von BinderHub) ist eine Open-Source-Lösung für das freie zur Verfügungstellung von Jupyter Notebooks. Sobald ein bestimmtes Repository im Webbrowser aufgerufen ist, wird eine komplette Ausführumgebung im Webbrowser gestartet.

### 3.1. Die Zellen Logik
Jupyter Notebooks arbeiten mit sogenannten Zellen. In Zellen kann Code oder Text stehen (zum Text kommen wir später). Damit eine Zelle ausgeführt wird, muss die Zelle angewählt sein und im Anschluss der Run-Knopf im der Befehlszeile oder die Tastenkombination **ctrl + Enter** gedrückt werden.

*Tipps:* 
- Mit der Tastenkombination **ctrl + Enter** wird eine Zelle ausgeführt.
- Mit **Shift + Enter** wird die Zelle ausgeführt und die nächste Zelle ausgewählt.
- Mit **Alt + Enter** wird die Zelle ausgeführt und als nächstes eine neue Zelle erstellt.
- Mit **Esc + b** wird eine neue Zelle unterhalb und mit **Esc + a** eine Zelle oberhalb der ausgewählten Zelle erstellt.
- In der **Befehlszeile** befinden sich weitere Funktionen, z.B. eine Zelle ausschneiden, alle Zellen von oben nach unten ausführen, usw.


> 1. Führe die folgende Zelle aus.
