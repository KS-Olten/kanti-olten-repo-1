# 5. Bart II

... und wieder ist es passiert. Bart hat dieses Mal vergessen, den Tag zu geniessen. Zur Strafe soll er 100 Mal "Ich soll den Tag geniessen!" schreiben. Weil er es sich beim letzten Mal so einfach gemacht hat, hat sich die Lehrperson einen Trick ausgedacht: Bart soll die Zeilen nummerieren, damit er nicht einfach eine Schleife verwenden kann.


>Zeile 1: Ich soll den Tag geniessen!
>
>Zeile 2: Ich soll den Tag geniessen!
>
>Zeile 3: Ich soll den Tag geniessen!
>
>...


Bart vermutet, dass er das mit Variablen lösen kann. Allerdings kriegt er es nicht alleine hin. Kannst du ihm mit deinem Wissen helfen?

>>**Tipp:** Der Befehl "repeat [zahl]:" ist kein Standardbefehl von Python. Er ist spezifisch für TigerJython. In python wird folgender Befehl verwendet:
>>```python
>>for [variable_name] in range([zahl]):
>>```
>>
>>Dabei muss `[variable_name]` mit dem Namen einer Variable und `[zahl]` mit einer tatsächlichen Zahl ersetzt werden. Der eingerückte Code wird dann entsprechend der Zahl wiederholt, wobei die Variable die Werte 0 bis zur gewählten Zahl durchschreitet.

````python
zeile = 1
for i in range(100):
    print("Zeile", zeile, "Ich soll den Tag geniessen!")
    zeile = zeile + 1
    # Mötlich wäre auch: print("Zeile", i+1, "Ich soll den Tag geniessen!")
````

Mit dem Befehl

```python

```

Kann einer Variable ein Nutzer-Input gegeben werden. Führe die folgende Code-Zelle aus und du wirst sehen, wie es funktioniert.

>**a)** Passe den Code so an, dass der Nutzer\*in nicht nur das Alter in Jahren zurückgegeben wird, sondern auch in Sekunden, Tagen, Wochen und Monaten.

>**b)** Erstelle eine weitere Code-Zelle und erstelle nochmals einen Alters-Rechner. Dieser Rechner soll aber zurückgeben, wie viele Jahre, Monate und Tage die Nutzer\*in noch zu leben hat. Wir nehmen eine durchschnittliche Lebenserwartung ovon 85 Jahren an.

>>**Tipp:** Wemm du mit der Variable `jahre` rechnen möchtest, wirst du sehr wahrscheinlich in einem ersten Moment Probleme haben. Bedenke, dass die `input()`-Methode den eingegebenen Text dynamisch als String erkennt...

````python
print("Dein Alter in Jahren:")

jahre = input()
monate = int(jahre)*12
wochen = int(jahre)*52
tage = int(jahre)*365

#Bonus
sekunden = int(jahre)*365*24*60*60


print("Jahre:", jahre)
print("Monate:", monate)
print("Wochen:", wochen)
print("Tage:", tage)
#Bonus
print("Sekunden:", sekunden)
````

Wird ein *Canvas*, also sozusagen ein Bildrahmen erstellt. Der Bildrahmen besteht aus einem Koordinatensystem, auf welches du Formen zeichnen kannst. Der folgende Code zeichnet zum Beispiel zwei Quadrate. Das darauf folgende Bild zeigt, wie die Positionierung der Quadrate zustande kommt.

````python
from ipycanvas import Canvas                        #Importiere Canvas-Modul

canvas_size = 500                                   #Lege Dimension des Canvas fest
canvas = Canvas(size=(canvas_size, canvas_size)) 
x = 10
y = 20
width = 50
height = 100

canvas.fill_style = 'green'
canvas.fill_rect(x, y, width, height)

canvas.fill_style = 'red'
x = x + 300
y = y + 200
canvas.fill_rect(x, y, width, height)

canvas                      
````

**Aufgabe:** Schaffst du es, ein solches Bild mit einer Schleifen und Variablen zu zeichnen?

````python
from ipycanvas import Canvas                        #Importiere Canvas-Modul

canvas_size = 500                                   #Lege Dimension des Canvas fest
canvas = Canvas(size=(canvas_size, canvas_size))
x = 0
y = 0
width = 50
height = 50

for i in range(10):
    

    canvas.fill_style = 'green'
    canvas.fill_rect(x, y, width, height)
    
    x_2 = canvas_size - x - width
    canvas.fill_rect(y, x_2, width, height)
    
    x = x + width
    y = y + width
    
    canvas.fill_style = 'red'
    canvas.fill_rect(x, y, width, height)
    
    x_2 = canvas_size - x - width
    canvas.fill_rect(y, x_2, width, height)
    
    x = x + width
    y = y + width


canvas    
````
